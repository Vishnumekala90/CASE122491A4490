```python
!pip install easyocr
!pip install imutils

import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
import easyocr

img = cv2.imread("/content/ch.jpg")

# Step 2: Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes
plt.title("Grayscale Image")
plt.show()

# Step 3: Apply filter and find edges for localization
bfilter = cv2.bilateralFilter(gray, 11, 17, 17)
edged = cv2.Canny(bfilter, 30, 200)


# Find contours
keypoints = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(keypoints)
contours = sorted(contours, key=cv2.contourArea, reverse=True)[:10]

location = None
for contour in contours:
    approx = cv2.approxPolyDP(contour, 10, True)
    if len(approx) == 4:
        location = approx
        break

# Check if location was found
if location is not None:
    # Create a mask
    mask = np.zeros(gray.shape, dtype=np.uint8)  # Corrected shape parameter
    new_image = cv2.drawContours(mask, [location], -1, 255, -1)  # -1 to draw all contours
    new_image = cv2.bitwise_and(img, img, mask=mask)

    # Display the new image
    plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Hide axes
    plt.title("Contour Image")
    plt.show()

    # Get the coordinates of the mask
    (x, y) = np.where(mask == 255)
    (x1, y1) = (np.min(x), np.min(y))
    (x2, y2) = (np.max(x), np.max(y))
    cropped_image = gray[x1:x2 + 1, y1:y2 + 1]  # Fixed slicing

    # Optionally display the cropped image
    plt.imshow(cropped_image, cmap='gray')
    plt.axis('off')  # Hide axes
    plt.title("Cropped Image")
    plt.show()
else:
    print("No quadrilateral contour found.")


# Step 4: Use EasyOCR to read text
reader = easyocr.Reader(['en'])

# Read text from the cropped image
result = reader.readtext(cropped_image)

# Check if any text was found
if result:
    for (bbox, text, prob) in result:
        print(f"Detected text: {text} with confidence: {prob:.2f}")

    # Optionally, you can visualize the results on the image
    for (bbox, text, prob) in result:
        # Draw bounding box
        cv2.rectangle(cropped_image,
                      (int(bbox[0][0]), int(bbox[0][1])),
                      (int(bbox[2][0]), int(bbox[2][1])),
                      (0, 255, 0), 2)
        # Put text
        cv2.putText(cropped_image, text,
                    (int(bbox[0][0]), int(bbox[0][1]) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the cropped image with detected text
    plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Hide axes
    plt.title("Detected Text in Cropped Image")
    plt.show()
else:
    print("No text detected.")

if result:
    text = result[0][-2]  # Extract the detected text
    font = cv2.FONT_HERSHEY_SIMPLEX  # Corrected font constant

    # Draw the text on the image
    res = cv2.putText(img, text=text,
                      org=(int(approx[0][0][0]), int(approx[0][0][1] + 60)),
                      fontFace=font, fontScale=1,  # Corrected argument name
                      color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)

    # Draw rectangle around the detected text area
    res = cv2.rectangle(res,
                        tuple(map(int, approx[0][0])),  # Top-left corner
                        tuple(map(int, approx[2][0])),  # Bottom-right corner
                        (0, 255, 0), 3)  # Color and thickness

    # Display the resulting image
    plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Hide axes
    plt.title("Detected Text with Rectangle")
    plt.show()
else:
    print("No text detected.")
```
