```python
import pandas as pd
import matplotlib.pyplot as plt
dt=pd.read_csv("/content/archive (2).zip")
#print(dt)
print(dt.head(2))
print(dt.tail(3))
print(dt.shape)
print(dt.columns)
print(dt.dtypes)
print(dt.describe())
print(dt.info())
us=dt[dt["university"]=="PU"]
ap=dt[dt["university"]=="UTE"]
fig,diag=plt.subplots(2)
diag[0].plot(us["cgpa"],us["average_sleep"])
diag[1].plot(ap["cgpa"],ap["average_sleep"])
ts=dt[dt["degree_major"]=="Computer Science"]
up=dt[dt["degree_major"]=="Data Science"]
fig,diag=plt.subplots(2)
diag[0].bar(ts["financial_concerns"],ts["depression"])
diag[1].bar(up["financial_concerns"],up["depression"])
vi="Religious Activities, Social Connections, Online Entertainment"
sh="Online Entertainment"
nu=(vi,sh)
me=dt[dt["stress_relief_activities"]=="Religious Activities, Social Connections, Online Entertainment"]["future_insecurity"].sum()
ka=dt[dt["stress_relief_activities"]=="Online Entertainment"]["future_insecurity"].sum()
la=(me,ka)
fig,diag=plt.subplots(2)
diag[0]=plt.pie(la,labels=nu)
```
