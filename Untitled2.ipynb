```python
import pandas as pd
import matplotlib.pyplot as plt
hi=pd.read_csv('energy.csv')
print(hi)
srh=hi[hi['State']=='Andhra']
rcb=hi[hi['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])
srh=hi[hi['State']=='Telengana']
rcb=hi[hi['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'])
diag[1].bar(rcb['Date'],rcb['Consumption'])

csds='Kerala'
cse='Delhi'
ece=[csds,cse]

tdp=hi[hi['State']=='Kerala']['Consumption'].sum()
ysr=hi[hi["State"]=="Delhi"]["Consumption"].sum()
jsp=[tdp,ysr]
fig,diag=plt.subplots(2)
diag[0].pie(jsp,labels=ece,autopct='%1.1f%%')
```
